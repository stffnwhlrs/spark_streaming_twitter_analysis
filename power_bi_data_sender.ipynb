# Puplic opinion sender

```python
def send_public_opinion(company, tweet_count, sentiment_positive, time):
  # REST address to PowerBI
  url = "https://api.powerbi.com/beta/73458443-1627-4091-8b39-2222134907c5/datasets/20792726-b470-438e-a12c-6c99b9bba033/rows?key=p9e7FOz8G%2FkXhiHCGQ1rzm1MydzL8r26XyMdMcFcRMFWY4h6sUv4uatSqNQeR36ZC0%2FFDNwTbxsFJzboM57bZg%3D%3D"

  # The data that shall be send to PowerBI
  data = [
      {
      "company" :company,
      "tweet_count" :tweet_count,
      "sentiment_positive" :sentiment_positive,
      "sentiment_negative" :1 - sentiment_positive,
      "time" : time, 
      "misc_text" :"",
      "misc_num" :0
      }
  ]

  # Send data to PowerBI
  response = requests.post(url,json=data)
  # Print successfull. Success = 200
  print(response)
```
